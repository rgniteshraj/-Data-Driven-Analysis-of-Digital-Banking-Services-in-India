# Data Cleaning Steps



## Stage 2 – Data Cleaning and Pre-processing

After performing the initial exploratory analysis in Stage 1, the dataset was cleaned and prepared for further analysis. The following preprocessing steps were carried out:


1.Handle Missing Values

Missing values were already identified during Stage 1. In this stage, the missing values were treated appropriately to ensure data quality and consistency.




```python
# Fill numerical missing values with median
num_cols = df.select_dtypes(include=['int64','float64']).columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())

# Some columns had a few missing values. These missing values were replaced with the median value of the column to maintain data quality and avoid affecting the analysis.
```

2.Handle Duplicate Records

During Stage 1, duplicate records were checked using df.duplicated().sum() and no duplicate records were found in the dataset. Therefore, no duplicate removal was required.

```python
  # No duplicate records found, so no removal was required.
```

3.Convert Data Types

The reporting date column was converted into datetime format for time-series analysis.

```python
# Convert date column
df['date'] = pd.to_datetime(df['date'])
```

```python
df.info()
```

4.Create Derived Features

New features were created to enhance analysis

```python
# Total ATM Network
df['total_atms'] = df['atms_crms_onsite'] + df['atms_crms_offsite']

# Total Cards Issued
df['total_cards'] = df['credit_cards'] + df['debit_cards']

# Total Online Transaction Value
df['total_online_value'] = (
    df['cc_pay_trns_online_val'] +
    df['dc_pay_trns_online_val']
)

# Total POS Transaction Value
df['total_pos_value'] = (
    df['cc_pay_trns_at_pos_val'] +
    df['dc_pay_trns_at_pos_val']
)
```

total_atms → Total ATM network of a bank.

total_cards → Total cards issued by a bank.

total_online_value → Combined online transaction value from debit and credit cards.

total_pos_value → Combined POS transaction value from debit and credit cards.

5.Verify Cleaned Dataset

```python
print("Dataset Shape:", df.shape)
print("Missing Values:", df.isnull().sum().sum())
print("Duplicate Records:", df.duplicated().sum())
```

6.Save Cleaned Dataset

```python
df.to_csv('cleaned_banking_dataset.csv', index=False)
```

##Stage 3 – Exploratory Data Analysis and Visualizations

The purpose of Stage 3 is to perform Exploratory Data Analysis (EDA) using visualizations to identify patterns, trends, relationships, and business insights from the cleaned banking dataset.

Visualization 1: Top 10 Banks by Total ATM Network

Purpose : To identify the top 10 banks with the largest ATM network and compare their banking infrastructure.

Why I Used This Chart:

I used a bar chart because it is effective for comparing values across multiple banks, and long bank names are easier to read.

```python
# Group the data from bank name and calculate the total Atms
top_atm = df.groupby('bank_name')['total_atms'].sum()
top_atm = top_atm.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,5))
top_atm.plot(kind='barh')
plt.title('Top 10 Banks by ATM Network')
plt.xlabel('Total ATMs')
plt.ylabel('BRANCH NAME')
plt.show()
```

Interpretation

This bar chart shows the top 10 banks with the highest number of ATMs. It helps compare the ATM infrastructure of different banks. Banks with more ATMs generally have a wider physical presence and can provide better access to banking services for customers.


Visualization 2: Top 10 Banks by POS Terminals

Purpose : To compare the top 10 banks by POS terminals in a cleaner and more visually appealing way.

Why I Used This Chart:

I used a lollipop chart because it provides a clean and simple comparison between banks while using less visual space than a bar chart.

```python

top_pos = df.groupby('bank_name')['pos'].sum()

top_pos =top_pos.sort_values(ascending=False).head(10)
plt.figure(figsize=(10,6))

plt.hlines(
    y=top_pos.index,
    xmin=0,
    xmax=top_pos.values
)

plt.plot(
    top_pos.values,
    top_pos.index,
    "o"
)

plt.title("Top 10 Banks by POS Terminals")

plt.xlabel("Number of POS Terminals")
plt.ylabel("Bank Name")
plt.show()
```

Interpretation

This lollipop chart compares the top 10 banks by POS terminals. Banks with more POS terminals have a stronger merchant payment infrastructure, supporting greater adoption of cashless transactions.

Visualization 3: Debit Cards vs Credit Cards

Purpose : To compare the total number of debit cards and credit cards issued by banks and understand customer preference for different card types.

Why I Used This Chart:

I used a pie chart because it clearly displays the percentage share of each card type

```python
# Calculate totals
debit = df['debit_cards'].sum()
credit = df['credit_cards'].sum()

# Create pie chart
plt.pie(
    [debit, credit],
    labels=['Debit Cards', 'Credit Cards'],
    autopct='%1.1f%%'
)

plt.title('Debit Cards vs Credit Cards')
plt.show()
```

Interpretation

This  pie chart compares the overall card distribution in India. It helps understand whether customers use debit cards or credit cards more frequently.

Visualization 4: Bank Category Distribution

Purpose : To understand the distribution of different categories of banks in the dataset.

Why I Used This Chart:

I used a count plot because it automatically counts and compares the frequency of each bank category.

```python
# Create a count plot for each bank category
sns.countplot(data=df, x='bank_category')
plt.xticks(rotation=45)
plt.title('Bank Category Distribution')
plt.xlabel('Bank Category')
plt.ylabel('Count')
plt.show()
```

 Interpretation

This chart shows the number of banks in each category and helps understand the composition of the banking sector.



Visualization 5: Online Transaction Value Trend

Purpose: To analyze how online transaction values change over time and identify trends in digital banking.

Why I Used This Chart:

I used a line chart because it is the most suitable chart for showing trends and changes over time.

```python
# Group data by date and calculate total online transaction value
monthly = df.groupby('date')['total_online_value'].sum()

plt.figure(figsize=(12,5))
monthly.plot()
plt.title('Online Transaction Value Trend')
plt.xlabel('Date')
plt.ylabel('Transaction Value')
plt.show()
```

Interpretation

This line chart shows the trend in online transaction values and indicates the growth of digital banking over time

Visualization 6 – Online Transaction Value by Bank Category

Purpose

To compare the distribution of online transaction values across different bank categories and identify variations in digital banking performance.

Why I Used This Chart:

I used a box plot because it shows the median, spread, and outliers of transaction values, making it easy to compare distributions

```python

plt.figure(figsize=(10,6))
# Create a box plot
sns.boxplot(
    data=df,
    x='bank_category',
    y='total_online_value'
)
# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.title('Online Transaction Value by Bank Category')
plt.xlabel('Bank Category')
plt.ylabel('Online Transaction Value')
plt.show()
```

Interpretation

The box plot compares the distribution of online transaction values across different bank categories. It shows the median, spread, and possible outliers for each category. Categories with higher median values generally have greater digital banking activity, while wider spreads indicate greater variation in transaction values among banks

Visualization 7: UPI QR Adoption

Purpose : To compare the distribution of UPI QR deployment across different bank categories.

Why I Used This Chart:
I used a violin plot because it displays both the distribution and density of the data, providing more information than a box plot.

```python

plt.figure(figsize=(10,6))
# Create violin plot
sns.violinplot(
    data=df,
    x='bank_category',
    y='upi_qr'
)
# Rotate x-axis labels
plt.xticks(rotation=45)
plt.title('UPI QR Distribution by Bank Category')
plt.xlabel('Bank Category')
plt.ylabel('UPI QR Count')

plt.show()
```

Interpretation

This Violin charts shows the distribution and density of UPI QR counts across different bank categories. It highlights the spread, concentration, and variation in UPI QR deployment, helping compare digital payment infrastructure among bank categories.

Visualization 8: ATM vs POS Relationship

Purpose : To examine the relationship between ATM infrastructure and POS terminal deployment across banks.

Why I Used This Chart:
I used a scatter plot because it is suitable for analyzing the relationship between two numerical variables.

```python

plt.figure(figsize=(10,5))
# Create a scatter plot between ATM network and POS terminals
sns.scatterplot(
    data=df,
    x='total_atms',
    y='pos'
)
# Add chart title
plt.title('ATM Network vs POS Terminals')
# Add X-axis label
plt.xlabel('Total ATMs')
# Add Y-axis label
plt.ylabel('POS Terminals')
# Display the chart
plt.show()
```

Interpretation

The scatter plot shows a positive relationship between ATM networks and POS terminals. Banks with larger ATM infrastructure generally also have more POS terminals, although a few banks differ from this trend.

Visualization 9: Correlation Heatmap

Purpose : To identify relationships between numerical variables such as ATM count, POS count, card count, and transaction values.

Why I Used This Chart:
I used a heatmap because it visually represents correlation values using colors, making relationships easier to identify.

```python

plt.figure(figsize=(12,8))

# Calculate correlation among numerical columns
correlation = df.select_dtypes(include='number').corr()

# Draw the heatmap
sns.heatmap(
    correlation,
    annot=True,
    cmap='coolwarm'
)
##cmap='coolwarm' applies a color scheme to the heatmap.
##Red → Strong positive correlation (values close to +1)
##Blue → Strong negative correlation (values close to -1)
##Light colors → Weak or no correlation (values close to 0)

plt.title('Correlation Heatmap')

plt.show()
```

Interpretation

The correlation heatmap shows that most banking variables have a strong positive correlation, especially ATM network, total cards, POS terminals, and transaction values. This indicates that banks with larger banking infrastructure generally have higher digital transaction activity. Some "other transaction" variables show weak or negative correlations, indicating a limited relationship with the main banking metrics

Visualization 10: ATM Withdrawal Value Distribution

Purpose : To understand the distribution of debit card cash withdrawal values made through ATMs.

Why I Used This Chart:
I used a histogram because it shows how ATM withdrawal values are distributed across different ranges.

```python
plt.figure(figsize=(10,5))

# Plot the distribution of ATM withdrawal values
sns.histplot(
    df['dc_cash_withdraw_atm_val'],
    bins=20
)
plt.title('Distribution of ATM Withdrawal Values')
plt.xlabel('ATM Withdrawal Value')
plt.ylabel('Frequency')
plt.show()
```

Interpretation

The histogram shows that most ATM withdrawal values fall within a particular range, while high-value withdrawals are less frequent. This helps understand customer withdrawal patterns.
